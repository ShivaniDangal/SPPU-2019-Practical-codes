## Read the dataset

```python
import pandas as pd

df= pd.read_csv('/content/Churn_Modelling.csv')
df
```

## Distinguish the feature and target set and divide the data set into training and test sets

```python
cols= df.columns
cols
```

```python
x= df[['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']]
y= df['Exited']
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest= train_test_split(x,y, test_size=0.2, random_state=42)
```

## Normalize the train and test data

```python
from sklearn.preprocessing import MinMaxScaler

std= MinMaxScaler()
xtrain= std.fit_transform(xtrain)
xtest= std.transform(xtest)
```

## Initialize and build the model

```python
import tensorflow as tf
from tensorflow import keras

model= keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(xtrain.shape[1],)),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
model.fit(xtrain, ytrain, epochs=10, batch_size=32, validation_split=0.2)
```

```python
ypred= (model.predict(xtest)>0.5)
```

## Print the accuracy score and confusion matrix

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
cm= confusion_matrix(ytest, ypred)
cm
```

```python
accuracy= accuracy_score(ytest, ypred)
print(accuracy)
```
