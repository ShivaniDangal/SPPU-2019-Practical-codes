# Boxplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig,axs= plt.subplots(2,2,figsize=(16,9))
sns.boxplot(data=df, x='Species', y='SepalLengthCm', ax= axs[0,0])
sns.boxplot(data=df, x='Species', y='SepalWidthCm', ax= axs[0,1])
sns.boxplot(data=df, x='Species', y='PetalLengthCm', ax= axs[1,0])
sns.boxplot(data=df, x='Species', y='PetalWidthCm', ax= axs[1,1])
```

# Outlier detection

```python
import numpy as np
```

```python
dfg= df.groupby('Species')
vir= (dfg.get_group('Iris-virginica'))
print(np.where(vir['SepalLengthCm']<5.5))
#print(np.where(df['SepalLengthCm']<5.5  df['Species']=='Iris-virginica'))
```

# Histogram

```python
fig,axs= plt.subplots(2,2,figsize=(16,9))
sns.histplot(data=df, x='SepalLengthCm', ax= axs[0,0])
sns.histplot(data=df, x='SepalWidthCm', ax= axs[0,1])
sns.histplot(data=df, x='PetalLengthCm', ax= axs[1,0])
sns.histplot(data=df, x='PetalWidthCm', ax= axs[1,1])
```
