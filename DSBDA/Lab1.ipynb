# Turn categorical variables into quantitative variables using technique: Label Encoder

```python
#convert categorical variable 'Species' into quantative variable
from sklearn import preprocessing

convert= preprocessing.LabelEncoder()
df['Species']= convert.fit_transform(df['Species'])
df
```
